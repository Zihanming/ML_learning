# 📏 Distance Metrics 简介

在机器学习和数据科学中，**距离度量（Distance Metric）** 是衡量两个样本（点、向量或集合）之间相似度或差异性的核心工具。选择合适的距离度量，能直接影响模型效果，尤其在以下场景中至关重要：

- 📍 **聚类算法（如 K-Means、DBSCAN）**
- 🔍 **最近邻搜索（如 KNN）**
- 🧠 **推荐系统（基于内容的推荐）**
- 📊 **降维与特征选择（如 PCA、LDA）**
- 🚨 **异常检测**

---

## ✳️ 什么是“好”的距离度量？

一个好的距离度量应该满足以下**度量公理**（Metric Axioms）：

1. **非负性（Non-negativity）**：$d(x, y) \geq 0$
2. **自反性（Identity of Indiscernibles）**：$d(x, y) = 0 \Leftrightarrow x = y$
3. **对称性（Symmetry）**：$d(x, y) = d(y, x)$
4. **三角不等式（Triangle Inequality）**：$d(x, z) \leq d(x, y) + d(y, z)$

---

## 🎯 不同类型的距离指标用途不同

| 类别             | 示例                  | 常见应用                       |
|------------------|-----------------------|-------------------------------|
| Lp范数类距离      | 欧几里得距离、曼哈顿距离   | 图像分析、KNN、回归等               |
| 角度/相似度类距离 | 余弦相似度、相关距离     | 文本、推荐系统、嵌入空间建模           |
| 统计/分布类距离   | Mahalanobis 距离等      | 多维异常检测、分布建模、判别分析等       |

1. ✅ 欧几里得距离（Euclidean Distance）

## 1. 欧几里得距离（Euclidean Distance）

### 🧭 介绍
欧几里得距离是最常见的距离度量方式，用来衡量两个点之间的“直线距离”。它是多维空间中两点之间最短的路径，符合人类对“距离”的直观理解。

### 📐 公式
对于两个点 $\mathbf{x} = (x_1, x_2, ..., x_n)$ 和 $\mathbf{y} = (y_1, y_2, ..., y_n)$，欧几里得距离定义为：

$$
d(\mathbf{x}, \mathbf{y}) = \sqrt{ \sum_{i=1}^{n} (x_i - y_i)^2 }
$$

它等价于 $L_2$ 范数：$\|\mathbf{x} - \mathbf{y}\|_2$。

### 🔍 几何理解
欧几里得距离在二维空间中是两点之间的直线长度，是**毕达哥拉斯定理**在高维空间中的延伸。  
可视化时，两点之间的连接线即为该距离。

### 🧠 特性
- ✅ 满足度量公理（非负性、对称性、三角不等式）
- 📏 属于 $L_2$ 范数
- ⚠️ 对离群点敏感（平方项放大差异）
- 🔧 通常需归一化数据（防止量纲影响）

### 🧪 常用于
- K-Nearest Neighbors（KNN）
- K-Means 聚类
- Embedding 空间距离计算
- 图像、文本相似度度量（已归一化）

### ⚖️ 优缺点

| ✅ 优点               | ⚠️ 缺点                          |
|---------------------|----------------------------------|
| 直观、易理解           | 高维空间中会出现“距离退化”现象        |
| 推导和实现简单         | 对不同量纲的数据不鲁棒，需要标准化处理  |
| 满足所有度量公理       | 对异常值敏感，易被极端值拉大差异       |

### 🎯 面试常见问题

1. **欧几里得距离与 L2 范数是什么关系？**  
   它本质上就是 L2 范数，表示两个向量差的长度，即 $\|\mathbf{x} - \mathbf{y}\|_2$。

2. **欧几里得距离是否满足度量公理？**  
   是的，它满足非负性、自反性、对称性和三角不等式，因此是合法的距离度量。

3. **高维空间下欧几里得距离的问题？**  
   在高维下会出现“距离退化”（即所有点之间距离相近），导致模型判别力下降。

4. **什么时候不适合使用欧几里得距离？**  
   当特征存在量纲不一致、极端值或高维稀疏时，欧几里得距离可能不稳定。

5. **欧几里得距离适合哪些任务？**  
   适合低维、数值型、归一化后的数据，常用于 KNN、KMeans 和图像相似度计算中。



```python
# 欧几里得距离计算与可视化

import numpy as np
import matplotlib.pyplot as plt

# 定义欧几里得距离函数
def euclidean_distance(x, y):
    return np.linalg.norm(x - y)

# 示例向量
x = np.array([1, 2])
y = np.array([4, 6])

# 距离计算
dist = euclidean_distance(x, y)
print(f"Euclidean Distance: {dist:.2f}")

# 可视化两点之间的欧几里得距离
plt.figure(figsize=(6, 6))
plt.scatter(*x, color='blue', label='x (1, 2)', s=100)
plt.scatter(*y, color='red', label='y (4, 6)', s=100)
plt.plot([x[0], y[0]], [x[1], y[1]], 'k--', label=f'Distance = {dist:.2f}')
plt.text((x[0]+y[0])/2 + 0.1, (x[1]+y[1])/2, f'{dist:.2f}', fontsize=12)
plt.title('X_Y distance')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```

## 2. 曼哈顿距离（Manhattan Distance）

### 🧭 介绍
曼哈顿距离，也称为“城市街区距离”或 $L_1$ 范数，是指两个点之间沿坐标轴方向的距离总和，就像在棋盘或城市道路上只能沿水平和垂直方向移动一样。

### 📐 公式
对于向量 $\mathbf{x} = (x_1, ..., x_n)$ 和 $\mathbf{y} = (y_1, ..., y_n)$，曼哈顿距离定义为：

$$
d(\mathbf{x}, \mathbf{y}) = \sum_{i=1}^{n} |x_i - y_i|
$$

即对应维度的差值的绝对值之和。

### 🔍 几何理解
曼哈顿距离可以想象成“只能沿横竖走”的路径长度。例如，从一个城市街区到另一个街区只能走马路而不能穿过建筑。

### 🧠 特性
- ✅ 满足度量公理
- 📏 属于 $L_1$ 范数
- 🚫 不强调方向，仅沿坐标轴分量求和
- ✅ 对异常值更鲁棒（不含平方项）

### 🧪 常用于
- L1 正则化模型（如 Lasso 回归）
- KNN（对高维鲁棒）
- 图像、文本、离散特征相似度计算
- 搜索空间为 grid 时（如路径规划）

### ⚖️ 优缺点

| ✅ 优点                    | ⚠️ 缺点                         |
|---------------------------|----------------------------------|
| 对离群点不敏感（无平方）     | 不能很好反映点与点的方向性关系         |
| 易于计算、无需开方            | 不适合需要“最短直线距离”的模型         |
| 在高维空间比 L2 更鲁棒       | 不连续可导，优化模型可能不光滑           |

### 🎯 面试常见问题（含简答）

1. **曼哈顿距离与欧几里得距离的核心区别？**  
   曼哈顿只考虑坐标轴方向移动，欧几里得是最短直线距离；一个是 L1 范数，一个是 L2 范数。

2. **L1 距离为何对异常值更鲁棒？**  
   因为不含平方项，对极大差异不会像欧几里得那样放大影响。

3. **曼哈顿距离是否满足度量公理？**  
   满足。它是合法的距离度量函数。

4. **Lasso 回归与曼哈顿距离的关系？**  
   Lasso 回归的正则项就是 L1 范数，即曼哈顿距离，使得特征稀疏性增强。

5. **使用场景中为何 Manhattan 更适合离散空间？**  
   在路径只能沿网格移动（如 A* 算法、路径规划）时，曼哈顿距离更加真实贴近实际。


```python
# 📌 曼哈顿距离计算与可视化

import numpy as np
import matplotlib.pyplot as plt

# 定义曼哈顿距离函数
def manhattan_distance(x, y):
    return np.sum(np.abs(x - y))

# 示例向量
x = np.array([1, 2])
y = np.array([4, 6])

# 计算距离
dist = manhattan_distance(x, y)
print(f"Manhattan Distance: {dist:.2f}")

# 可视化路径（沿 x 轴，再沿 y 轴）
plt.figure(figsize=(6, 6))
plt.scatter(*x, color='blue', label='x (1,2)', s=100)
plt.scatter(*y, color='red', label='y (4,6)', s=100)
plt.plot([x[0], y[0]], [x[1], x[1]], 'k--')  # 横向移动
plt.plot([y[0], y[0]], [x[1], y[1]], 'k--')  # 纵向移动
plt.text((x[0]+y[0])/2, x[1] - 0.5, f'Δx = {abs(x[0]-y[0])}', fontsize=10)
plt.text(y[0] + 0.2, (x[1]+y[1])/2, f'Δy = {abs(x[1]-y[1])}', fontsize=10)
plt.title(f'Manhattan Distance: {dist:.2f}')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()
```

## 3. 余弦相似度（Cosine Similarity）

### 🧭 介绍
余弦相似度用于衡量两个向量之间的“方向”是否相近，而非“距离”大小。它关注两个向量之间的夹角，广泛用于文本相似度、推荐系统、图像检索等方向敏感的任务中。

### 📐 公式
设 $\mathbf{x}, \mathbf{y}$ 为两个向量，则余弦相似度定义为：

$$
\cos(\theta) = \frac{\mathbf{x} \cdot \mathbf{y}}{\|\mathbf{x}\| \|\mathbf{y}\|}
$$

其取值范围为 $[-1, 1]$，其中：
- $1$ 表示方向完全相同
- $0$ 表示正交（无相关性）
- $-1$ 表示方向完全相反

有时我们使用 **余弦距离** 作为度量：$1 - \cos(\theta)$

### 🔍 几何理解
余弦相似度反映了两个向量之间的夹角，**不考虑长度大小**。即使一个向量是另一个的数倍，只要方向一致，相似度为 1。

### 🧠 特性
- ✅ 关注方向而非模长
- ✅ 不受特征量纲影响（默认归一化）
- ❌ 不满足三角不等式，因此**不是严格意义的距离度量**

### 🧪 常用于
- NLP 中的句子/文档相似度（TF-IDF、BERT）
- 推荐系统中的物品/用户相似性
- 图像检索和嵌入空间分析

### ⚖️ 优缺点

| ✅ 优点                          | ⚠️ 缺点                         |
|----------------------------------|----------------------------------|
| 对特征大小不敏感，适合高维稀疏向量   | 无法捕捉模长差异，例如频率大小      |
| 易于归一化后比较方向             | 不满足三角不等式，非严格距离        |
| 适合 TF-IDF、嵌入向量           | 不适合量纲相同但模长含义重要的任务  |

### 🎯 面试常见问题（含简答）

1. **余弦相似度和欧几里得距离的最大区别？**  
   欧几里得关注“大小+方向”，余弦只关注方向（夹角）。

2. **余弦相似度为什么适合文本？**  
   文本向量通常是高维稀疏，频率差异大但方向一致性重要，适合使用归一化后对比方向。

3. **为什么说余弦“不是严格的距离度量”？**  
   它不满足三角不等式，因此不是数学意义上的 metric。

4. **高维空间中，为什么使用余弦优于欧几里得？**  
   欧几里得距离易退化（差异趋于一致），余弦相似度更稳定，能保持角度信息。

5. **什么情况下不能用余弦相似度？**  
   当向量长度本身具有物理或统计含义（如收入、距离）时，不应忽略模长。


```python
# 📌 余弦相似度计算与可视化

import numpy as np
import matplotlib.pyplot as plt

# 定义余弦相似度函数
def cosine_similarity(x, y):
    dot_product = np.dot(x, y)
    norm_product = np.linalg.norm(x) * np.linalg.norm(y)
    return dot_product / norm_product

# 示例向量
x = np.array([1, 1])
y = np.array([2, 0])

# 计算相似度
sim = cosine_similarity(x, y)
angle = np.arccos(sim) * 180 / np.pi
print(f"Cosine Similarity: {sim:.3f}（Angle {angle:.1f}°）")

# 可视化向量夹角
plt.figure(figsize=(6, 6))
origin = [0], [0]
plt.quiver(*origin, x[0], x[1], angles='xy', scale_units='xy', scale=1, color='blue', label='x')
plt.quiver(*origin, y[0], y[1], angles='xy', scale_units='xy', scale=1, color='red', label='y')
plt.xlim(-1, 3)
plt.ylim(-1, 3)
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')
plt.title(f'Cosine Similarity = {sim:.3f}, Angle ≈ {angle:.1f}°')
plt.legend()
plt.show()
```

## 4. 马氏距离（Mahalanobis Distance）

### 🧭 介绍
马氏距离是基于协方差矩阵的距离度量方法，考虑了各特征之间的相关性和分布尺度。它常用于**异常检测**、**判别分析**等任务中，是“标准化后的多维欧几里得距离”。

### 📐 公式
对于样本向量 $\mathbf{x}$ 和均值向量 $\boldsymbol{\mu}$，协方差矩阵为 $\mathbf{S}$，则马氏距离定义为：

$$
d_M(\mathbf{x}, \boldsymbol{\mu}) = \sqrt{(\mathbf{x} - \boldsymbol{\mu})^T \mathbf{S}^{-1} (\mathbf{x} - \boldsymbol{\mu})}
$$

若协方差矩阵为单位矩阵 $\mathbf{I}$，则马氏距离退化为欧几里得距离。

### 🔍 几何理解
马氏距离会根据特征之间的协方差对不同方向进行“拉伸”或“压缩”。高相关性的方向被压缩，低方差方向被放大，从而使得距离计算对数据分布更敏感。

### 🧠 特性
- ✅ 满足度量公理
- 📏 距离单位无量纲化、考虑了特征之间的相关性
- 🚫 需要样本足够估计协方差矩阵，维度过高时不稳定
- 🔁 可推广为多元高斯分布中的马氏投影

### 🧪 常用于
- 多维异常值检测（与均值距离远表示“异常”）
- 判别分析（如 LDA）
- 聚类中的相关特征重加权
- 图像相似度、医学诊断建模

### ⚖️ 优缺点

| ✅ 优点                              | ⚠️ 缺点                                      |
|--------------------------------------|---------------------------------------------|
| 考虑特征之间的相关性，避免维度偏倚     | 计算协方差矩阵和逆矩阵开销大                 |
| 单位无关，数据标准化后仍有效            | 高维数据易造成矩阵不可逆或病态                |
| 适合用于高斯分布建模、异常检测           | 不适用于非线性或极非高斯分布的数据            |

### 🎯 面试常见问题（含简答）

1. **马氏距离与欧几里得距离的核心区别？**  
   欧几里得忽略特征相关性，马氏考虑协方差结构并进行归一化。

2. **协方差矩阵在马氏距离中起什么作用？**  
   它将数据从“原空间”投影到“无相关的标准空间”中再度量距离。

3. **马氏距离在哪些模型中常用？**  
   LDA、异常检测、多元高斯建模中非常常见。

4. **什么情况下不能使用马氏距离？**  
   当样本量小而维度高（n < d）时，协方差矩阵不可逆或估计不稳定。

5. **马氏距离是否满足度量公理？是否为真正距离？**  
   是的，马氏距离满足四个度量公理，是严格意义上的距离函数。


```python
# 📌 马氏距离计算与二维可视化

import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import distance

# 生成二维样本点集（均值 + 协方差结构）
np.random.seed(0)
mu = np.array([0, 0])
cov = np.array([[3, 1], [1, 2]])
X = np.random.multivariate_normal(mu, cov, size=100)

# 要计算马氏距离的点
x_query = np.array([2, 2])

# 计算协方差矩阵和其逆
S = np.cov(X.T)
S_inv = np.linalg.inv(S)

# 马氏距离计算函数
def mahalanobis_distance(x, mean, S_inv):
    diff = x - mean
    return np.sqrt(np.dot(np.dot(diff.T, S_inv), diff))

# 计算距离
mean_vec = np.mean(X, axis=0)
d = mahalanobis_distance(x_query, mean_vec, S_inv)
print(f"Mahalanobis Distance: {d:.3f}")

# 可视化点与数据分布
plt.figure(figsize=(6, 6))
plt.scatter(X[:, 0], X[:, 1], alpha=0.5, label='sample')
plt.scatter(*x_query, color='red', label='test', s=100)
plt.scatter(*mean_vec, color='black', marker='x', label='Ave', s=100)
plt.plot([mean_vec[0], x_query[0]], [mean_vec[1], x_query[1]], 'k--', label='direction')
plt.title(f'Mahalanobis Distance = {d:.3f}')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```

## 5. 相关系数距离（Correlation Distance）

### 🧭 介绍
相关系数距离是由皮尔逊相关系数推导出的相似性度量，反映的是两个向量之间的线性相关性。它不是关注“距离有多远”，而是关注两个变量是否“成正比”或“线性关系一致”。

### 📐 公式
设 $\mathbf{x}, \mathbf{y}$ 为两个向量，皮尔逊相关系数为：

$$
\rho(\mathbf{x}, \mathbf{y}) = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum (x_i - \bar{x})^2} \sqrt{\sum (y_i - \bar{y})^2}}
$$

相关系数距离定义为：

$$
d_{corr}(\mathbf{x}, \mathbf{y}) = 1 - \rho(\mathbf{x}, \mathbf{y})
$$

范围通常为 $[0, 2]$，但如果 $\rho \in [-1, 1]$，则距离为 $[0, 2]$

### 🔍 几何理解
它关注的是**向量的中心化方向是否一致**，即**去均值后的夹角**。如果两个向量只是尺度不同或平移不同，但趋势一致，则它们“相关性强”。

### 🧠 特性
- 🚫 不满足三角不等式，不是严格距离
- ✅ 去均值 → 强调线性关系
- 📏 与余弦相似度关系紧密，但前者对位置敏感度更低

### 🧪 常用于
- 聚类（如使用 `scipy` 中 correlation metric）
- 特征选择（高相关性→冗余特征）
- 金融、基因组、时间序列中变量间关联分析

### ⚖️ 优缺点

| ✅ 优点                          | ⚠️ 缺点                           |
|----------------------------------|-----------------------------------|
| 衡量变量趋势一致性（非仅量级）       | 仅反映线性相关，非线性无效             |
| 去均值后更健壮、无量纲              | 对非正态分布或极端值仍较敏感            |
| 适合高维/归一化前特征的聚类        | 不适用于实际距离建模，不可作为 metric    |

### 🎯 面试常见问题（含简答）

1. **相关系数距离与余弦相似度有何异同？**  
   都是方向性度量，但余弦不去均值，相关系数先中心化后再比方向。

2. **什么场景下使用相关系数距离比欧几里得更合适？**  
   当我们更关注趋势（如股票价格走势）而不是绝对值时。

3. **相关系数距离是否是严格意义上的距离函数？**  
   否。它不满足三角不等式，不能用于基于度量空间的算法（如 MDS）。

4. **如何判断两个特征之间存在冗余性？**  
   若相关系数绝对值高（距离小），表示两个特征冗余，应去除其一。

5. **相关系数为 0 意味着距离为多少？其解释？**  
   距离为 1，表示无线性相关性，两变量趋势完全不一致。


```python
# 📌 相关系数距离计算与简单可视化

import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import correlation

# 示例向量：趋势一致但数值不同
x = np.array([10, 20, 30, 40, 50])
y = np.array([1, 2, 3, 4, 5])      # 完全线性相关
z = np.array([5, 4, 3, 2, 1])      # 完全负相关

# 计算相关系数距离
d_xy = correlation(x, y)
d_xz = correlation(x, z)

print(f"相关系数距离 x vs y: {d_xy:.2f} （正相关 → 低距离）")
print(f"相关系数距离 x vs z: {d_xz:.2f} （负相关 → 高距离）")

# 可视化趋势
plt.figure(figsize=(8, 4))
plt.plot(x, label='x', marker='o')
plt.plot(y, label='y（Postive related）', marker='o')
plt.plot(z, label='z（Negative related）', marker='o')
plt.title('')
plt.legend()
plt.grid(True)
plt.show()
```
