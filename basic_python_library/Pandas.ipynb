### 3. æ•°æ®ç´¢å¼•ä¸é€‰å–ï¼ˆloc, iloc, at, iatï¼‰

#### âœ… `loc`: ä½¿ç”¨æ ‡ç­¾é€‰å–


````python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "Score": [85, 90, 95]
}, index=["a", "b", "c"])

# loc: æŒ‰æ ‡ç­¾é€‰å–
print("loc å•è¡Œï¼š\n", df.loc["a"])
print("loc å¤šåˆ—ï¼š\n", df.loc["a", ["Name", "Score"]])

# iloc: æŒ‰ä½ç½®é€‰å–
print("iloc ç¬¬2è¡Œç¬¬3åˆ—ï¼š", df.iloc[1, 2])  # Bob çš„ Score

# at: è·å–å•ä¸ªå…ƒç´ ï¼ˆæ ‡ç­¾ï¼‰
print("at æ–¹å¼ï¼š", df.at["b", "Age"])

# iat: è·å–å•ä¸ªå…ƒç´ ï¼ˆä½ç½®ï¼‰
print("iat æ–¹å¼ï¼š", df.iat[2, 1])  # Charlie çš„ Age
````

### 4. å¤„ç†ç¼ºå¤±å€¼ï¼ˆisna, dropna, fillnaï¼‰

åœ¨æ•°æ®åˆ†æä¸­ï¼Œç¼ºå¤±å€¼ï¼ˆNaN, nullï¼‰æ˜¯å¸¸è§é—®é¢˜ï¼ŒPandas æä¾›å¤šç§å‡½æ•°æ¥å¤„ç†ï¼š

| å‡½æ•°            | åŠŸèƒ½è¯´æ˜                     |
|-----------------|------------------------------|
| `isna()`        | åˆ¤æ–­ç¼ºå¤±å€¼ï¼Œè¿”å›å¸ƒå°”çŸ©é˜µ     |
| `notna()`       | åˆ¤æ–­éç¼ºå¤±å€¼                 |
| `dropna()`      | åˆ é™¤å«ç¼ºå¤±å€¼çš„è¡Œæˆ–åˆ—         |
| `fillna(x)`     | ç”¨æŒ‡å®šå€¼å¡«å……ç¼ºå¤±å€¼           |

---

#### `dropna()` å‚æ•°ï¼š

- `axis=0`ï¼šåˆ é™¤å«ç¼ºå¤±çš„è¡Œï¼ˆé»˜è®¤ï¼‰
- `axis=1`ï¼šåˆ é™¤å«ç¼ºå¤±çš„åˆ—
- `how="any"`ï¼ˆé»˜è®¤ï¼‰ï¼šåªè¦æœ‰ç¼ºå¤±å°±åˆ 
- `how="all"`ï¼šæ•´è¡Œ/æ•´åˆ—éƒ½ç¼ºå¤±æ‰åˆ 

---

#### `fillna()` å¸¸è§ç”¨æ³•ï¼š

- å›ºå®šå€¼å¡«å……ï¼š`fillna(0)`
- å‰å‘å¡«å……ï¼š`fillna(method="ffill")`
- åå‘å¡«å……ï¼š`fillna(method="bfill")`


````python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Name": ["Alice", "Bob", np.nan],
    "Age": [25, np.nan, 35],
    "Score": [85, 90, np.nan]
})

print("åŸå§‹æ•°æ®ï¼š\n", df)

# åˆ¤æ–­ç¼ºå¤±å€¼
print("\nisna åˆ¤æ–­ï¼š\n", df.isna())

# åˆ é™¤å«ç¼ºå¤±çš„è¡Œ
dropped = df.dropna()
print("\ndropna åï¼š\n", dropped)

# ç”¨ 0 å¡«è¡¥æ‰€æœ‰ç¼ºå¤±
filled = df.fillna(0)
print("\nfillna(0) å¡«è¡¥åï¼š\n", filled)

# å‰å‘å¡«å……ï¼ˆç”¨å‰ä¸€ä¸ªå€¼å¡«ï¼‰
ffill = df.fillna(method="ffill")
print("\nå‰å‘å¡«å…… ffillï¼š\n", ffill)

# åå‘å¡«å……ï¼ˆç”¨ä¸‹ä¸€ä¸ªå€¼å¡«ï¼‰
bfill = df.fillna(method="bfill")
print("\nåå‘å¡«å…… bfillï¼š\n", bfill)
````

### 5. æ•°æ®ç­›é€‰ä¸æ¡ä»¶è¿‡æ»¤ï¼ˆå¸ƒå°”ç´¢å¼•ï¼‰

Pandas å¯ä»¥é€šè¿‡å¸ƒå°”æ¡ä»¶æ¥é€‰å‡ºæ»¡è¶³æ¡ä»¶çš„è¡Œï¼ˆç±»ä¼¼ SQL ä¸­çš„ WHERE å­å¥ï¼‰ï¼š

#### åŸºæœ¬è¯­æ³•ï¼š

```python
df[æ¡ä»¶è¡¨è¾¾å¼]
```

#### å¤šæ¡ä»¶ç»„åˆï¼ˆå¸ƒå°”ç´¢å¼•ï¼‰ï¼š

åœ¨ Pandas ä¸­ç»„åˆå¤šä¸ªæ¡ä»¶æ—¶ï¼Œéœ€è¦ä½¿ç”¨ä½è¿ç®—ç¬¦ï¼ˆé€»è¾‘è¿ç®—ç¬¦ä¸èƒ½ç”¨ `and`, `or`, `not`ï¼‰ï¼š

- âœ… **ä¸**ï¼ˆANDï¼‰ï¼š`&`  
  è¯­æ³•ï¼š`(æ¡ä»¶1) & (æ¡ä»¶2)`  
  ç¤ºä¾‹ï¼š`(df["Age"] > 30) & (df["Score"] > 90)`

- âœ… **æˆ–**ï¼ˆORï¼‰ï¼š`|`  
  è¯­æ³•ï¼š`(æ¡ä»¶1) | (æ¡ä»¶2)`  
  ç¤ºä¾‹ï¼š`(df["Age"] < 25) | (df["Score"] < 80)`

- âœ… **é**ï¼ˆNOTï¼‰ï¼š`~`  
  è¯­æ³•ï¼š`~(æ¡ä»¶)`  
  ç¤ºä¾‹ï¼š`~df["Name"].isna()`

> ğŸ”’ **æ³¨æ„ï¼šæ¯ä¸ªæ¡ä»¶å¿…é¡»åŠ æ‹¬å·ï¼Œå¦åˆ™ä¼šæŠ¥é”™ï¼**



````python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Age": [25, 30, 35, 40],
    "Score": [85, 90, 95, 60]
})

# å•æ¡ä»¶ç­›é€‰ï¼šé€‰å‡ºå¹´é¾„å¤§äº 30 çš„äºº
print("Age > 30:\n", df[df["Age"] > 30])

# å¤šæ¡ä»¶ç­›é€‰ï¼šå¹´é¾„ > 30 ä¸”æˆç»© > 90
print("\nAge > 30 ä¸” Score > 90:\n", df[(df["Age"] > 30) & (df["Score"] > 90)])

# isinï¼šç­›é€‰ Name åœ¨ç‰¹å®šåˆ—è¡¨ä¸­çš„è¡Œ
print("\nName in ['Alice', 'Charlie']:\n", df[df["Name"].isin(["Alice", "Charlie"])])

# ~ éæ“ä½œç¬¦ï¼šæ’é™¤ç¼ºå¤±å€¼
df.loc[2, "Score"] = None
print("\néç¼ºå¤±å€¼ç­›é€‰ Score:\n", df[~df["Score"].isna()])
````

### 7. åˆ—æ“ä½œä¸æ–°å¢å­—æ®µï¼ˆå‘é‡è¿ç®—ã€assignã€applyï¼‰

---

#### âœ… å‘é‡è¿ç®—ï¼ˆæ¨èï¼‰

Pandas æ”¯æŒå¯¹æ•´åˆ—è¿›è¡Œå‘é‡åŒ–è¿ç®—ï¼Œé€Ÿåº¦å¿«ï¼Œè¯­æ³•ç®€å•ï¼š

```python
df["BMI"] = df["Weight"] / (df["Height"] / 100) ** 2
```

---

#### âœ… assign() æ–¹æ³•ï¼šé“¾å¼æ·»åŠ æ–°åˆ—ï¼ˆè¿”å›æ–° DataFrameï¼‰

```python
df = df.assign(ScoreRate=df["Score"] / 100)
```

é€‚åˆä¸ `.query()`ã€`.sort_values()` ç­‰é“¾å¼æ“ä½œç»„åˆä½¿ç”¨ã€‚

---

#### âœ… apply() æ–¹æ³•ï¼šæŒ‰è¡Œæˆ–åˆ—åº”ç”¨è‡ªå®šä¹‰å‡½æ•°

```python
df["Level"] = df["Score"].apply(lambda x: "High" if x > 90 else "Low")
```

- é»˜è®¤ `axis=0`ï¼šæŒ‰åˆ—ï¼ˆå¯¹æ¯ä¸ªâ€œå€¼åˆ—â€åº”ç”¨å‡½æ•°ï¼‰
- è‹¥ `axis=1`ï¼šæŒ‰è¡Œï¼ˆå¯¹æ¯ä¸€è¡Œä½œä¸ºæ•´ä½“ä¼ å…¥å‡½æ•°ï¼‰

---

#### âœ… è‡ªå®šä¹‰å‡½æ•° + applyï¼ˆæ›´å¤æ‚å¤„ç†ï¼‰

```python
def tag_level(row):
    if row["Score"] > 90:
        return "A"
    elif row["Score"] > 80:
        return "B"
    else:
        return "C"

df["Grade"] = df.apply(tag_level, axis=1)
```

> å‘é‡è¿ç®—é€Ÿåº¦æœ€å¿«ï¼›`apply` çµæ´»æ€§æœ€é«˜ï¼Œé€‚åˆå¤æ‚è§„åˆ™ã€‚


````python
import pandas as pd

# æ„é€ ç¤ºä¾‹æ•°æ®
df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie"],
    "Score": [95, 85, 72],
    "Weight": [60, 75, 68],     # kg
    "Height": [160, 180, 170]   # cm
})

# âœ… å‘é‡è¿ç®—ï¼šè®¡ç®— BMI = ä½“é‡ / (èº«é«˜ç±³)^2
df["BMI"] = df["Weight"] / (df["Height"] / 100) ** 2
print("å‘é‡è¿ç®—æ·»åŠ  BMI åˆ—ï¼š\n", df)

# âœ… assign æ·»åŠ æ–°åˆ—ï¼ˆé“¾å¼å†™æ³•ï¼‰
df = df.assign(ScoreRate = df["Score"] / 100)
print("\nä½¿ç”¨ assign æ·»åŠ  ScoreRate åˆ—ï¼š\n", df)

# âœ… apply + lambda åˆ›å»ºç­‰çº§æ ‡ç­¾
df["Level"] = df["Score"].apply(lambda x: "High" if x >= 90 else "Low")
print("\nä½¿ç”¨ apply + lambda åˆ›å»º Level åˆ—ï¼š\n", df)

# âœ… apply + è‡ªå®šä¹‰å‡½æ•°ï¼šæ ¹æ®åˆ†æ•°æ‰“ç­‰çº§
def tag_level(row):
    if row["Score"] > 90:
        return "A"
    elif row["Score"] > 80:
        return "B"
    else:
        return "C"

df["Grade"] = df.apply(tag_level, axis=1)
print("\nä½¿ç”¨ apply + è‡ªå®šä¹‰å‡½æ•° åˆ›å»º Grade åˆ—ï¼š\n", df)
````

### 8. åˆ†ç»„ä¸èšåˆåˆ†æï¼ˆgroupby, agg, pivot_tableï¼‰

Pandas æä¾›å¼ºå¤§çš„åˆ†ç»„ç»Ÿè®¡åŠŸèƒ½ï¼Œå¯æŒ‰æŸåˆ—å¯¹æ•°æ®è¿›è¡Œå½’ç±»å¤„ç†ï¼š

---

#### âœ… groupbyï¼šæŒ‰åˆ—åˆ†ç»„åèšåˆ

```python
df.groupby("åˆ—å").å‡½æ•°()
```

ç¤ºä¾‹ï¼š

```python
df.groupby("Department")["Salary"].mean()
```

---

#### âœ… aggï¼šä¸€æ¬¡å¯¹å¤šåˆ—æ‰§è¡Œå¤šä¸ªç»Ÿè®¡å‡½æ•°

```python
df.groupby("Department").agg({
    "Salary": "mean",
    "Age": ["max", "min"]
})
```

---

#### âœ… pivot_tableï¼šæ›´çµæ´»çš„é€è§†è¡¨ï¼ˆå¯èšåˆã€é€è§†ã€åˆ†å±‚ï¼‰

```python
pd.pivot_table(df, index="Department", values="Salary", aggfunc="mean")
```

ä¹Ÿæ”¯æŒå¤šé‡èšåˆå’Œå¤šä¸ªç»´åº¦ï¼š

```python
pd.pivot_table(df, index="Dept", columns="Gender", values="Salary", aggfunc="mean")
```

> `groupby` æ›´é€šç”¨ï¼Œ`pivot_table` æ›´é€‚åˆäºŒç»´ç»Ÿè®¡å±•ç¤º


````python
import pandas as pd

# ç¤ºä¾‹æ•°æ®ï¼šå…¬å¸å‘˜å·¥ä¿¡æ¯
df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank"],
    "Department": ["HR", "Tech", "Tech", "HR", "Tech", "Finance"],
    "Gender": ["F", "M", "M", "M", "F", "M"],
    "Salary": [5000, 8000, 7500, 5200, 8800, 6000],
    "Age": [25, 30, 35, 28, 32, 40]
})

# âœ… groupbyï¼šæŒ‰éƒ¨é—¨åˆ†ç»„ï¼Œç»Ÿè®¡å¹³å‡å·¥èµ„
print("å¹³å‡å·¥èµ„ï¼š\n", df.groupby("Department")["Salary"].mean())

# âœ… groupby + aggï¼šå¤šåˆ—å¤šå‡½æ•°èšåˆ
print("\næ¯ä¸ªéƒ¨é—¨çš„å¹³å‡å·¥èµ„ & æœ€å¤§å¹´é¾„ï¼š\n",
      df.groupby("Department").agg({
          "Salary": "mean",
          "Age": ["max", "min"]
      }))

# âœ… pivot_tableï¼šéƒ¨é—¨-æ€§åˆ«äº¤å‰è¡¨ï¼Œæ˜¾ç¤ºå¹³å‡å·¥èµ„
pivot = pd.pivot_table(df, index="Department", columns="Gender", values="Salary", aggfunc="mean")
print("\néƒ¨é—¨-æ€§åˆ« å¹³å‡å·¥èµ„é€è§†è¡¨ï¼š\n", pivot)
````

### 9. æ•°æ®åˆå¹¶ä¸è¿æ¥ï¼ˆmerge, concat, joinï¼‰

````python
import pandas as pd
import numpy as np

# æ„é€ æ—¶é—´åºåˆ—æ•°æ®ï¼šæ¯å¤©é”€é‡
date_range = pd.date_range(start="2023-01-01", periods=10, freq="D")
df = pd.DataFrame({
    "Date": date_range,
    "Sales": [10, 15, 13, 20, 23, 18, 30, 28, 25, 22]
})

# âœ… è½¬æ¢ä¸º datetime å¹¶è®¾ç½®ä¸ºç´¢å¼•
df["Date"] = pd.to_datetime(df["Date"])
df = df.set_index("Date")

# âœ… é‡é‡‡æ ·ï¼šæŒ‰æ¯ 3 å¤©æ±‚å¹³å‡
resampled = df.resample("3D").mean()
print("æ¯3å¤©å¹³å‡é”€é‡ï¼š\n", resampled)

# âœ… æ»šåŠ¨çª—å£ï¼š7 å¤©ç§»åŠ¨å¹³å‡
df["7d_avg"] = df["Sales"].rolling(window=7).mean()
print("\næ·»åŠ  7 å¤©æ»‘åŠ¨å¹³å‡åˆ—ï¼š\n", df)
````
